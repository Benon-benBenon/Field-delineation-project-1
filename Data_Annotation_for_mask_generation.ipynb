```python
import os
import json
import numpy as np
from PIL import Image, ImageDraw
from tqdm import tqdm
```

```python
def labelme_to_mask(json_path, output_path):
    with open(json_path, 'r') as f:
        data = json.load(f)

    width = data["imageWidth"]
    height = data["imageHeight"]

    mask = Image.new("L", (width, height), 0)
    draw = ImageDraw.Draw(mask)

    for shape in data["shapes"]:
        if shape["label"] == "Field":
            points = [(x, y) for x, y in shape["points"]]
            draw.polygon(points, fill=1)

    mask.save(output_path)
    print(f"✅ Saved: {output_path}")

# Example usage
json_folder = "C:/Users/NISR/Desktop/FIELD_DELINEATION & LAND_COVER/Annotated_not_very_cloudy"
output_folder = "C:/Users/NISR/Desktop/FIELD_DELINEATION & LAND_COVER/label_to_mask"
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(json_folder):
    if filename.endswith(".json"):
        json_path = os.path.join(json_folder, filename)
        output_name = filename.replace(".json", ".png")
        output_path = os.path.join(output_folder, output_name)
        labelme_to_mask(json_path, output_path)
```
